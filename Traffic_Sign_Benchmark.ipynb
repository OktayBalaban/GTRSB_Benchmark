```python
# Normalization

train_X = np.asarray(train_X)
train_X = train_X/255
train_X = np.asarray(train_X, dtype = "float32")
train_y = np.asarray(train_y, dtype= "float32")

test_X = np.asarray(test_X)
test_X = test_X/255
test_X = np.asarray(test_X, dtype = "float32")
test_y = np.asarray(test_y, dtype= "float32")

```

```python
# Validation Set

val_X = train_X[:12000]
val_y = train_y[:12000]
```

```python


model = keras.models.Sequential([
    keras.layers.Conv2D(32, kernel_size=5, strides=1,  activation='relu', input_shape=train_X[0].shape, padding='same'), #
    keras.layers.MaxPooling2D(), #
    keras.layers.BatchNormalization(),
    keras.layers.Conv2D(64, kernel_size=5, strides=1, activation='relu', padding='same'), #
    keras.layers.MaxPooling2D(), #
    keras.layers.BatchNormalization(),
    keras.layers.Flatten(), #Flatten
    keras.layers.Dense(196, activation='relu'), #
    keras.layers.BatchNormalization(),
    keras.layers.Dropout(.5, input_shape=(2,)),
    keras.layers.Dense(128, activation='relu'), #
    keras.layers.BatchNormalization(),
    keras.layers.Dropout(.5, input_shape=(2,)),
    keras.layers.Dense(96, activation='relu'), #
    keras.layers.BatchNormalization(),
    keras.layers.Dense(43, activation='linear') #with from_logits = True in optimizer, better than softmax
])
```

```python
model.compile(optimizer='Adam', loss=keras.losses.CategoricalCrossentropy(from_logits = True), metrics=['accuracy'])
```

```python
epoch = 0
maxAcc = 0
while True:
    epoch += 1
    model.fit(train_X, train_y, epochs=1, validation_data=(val_X, val_y), shuffle = True)
    score, acc = model.evaluate(test_X, test_y)
    if (acc > maxAcc):
        maxAcc = acc
    print('Epoch: {}'.format(epoch))
    print('Max Accuracy Reached: {maximum:.4f}'.format(maximum = maxAcc))
    print('Test Score: {loss:.4f}'.format(loss = score))
    print('Test Accuracy: {accuracy:4f}'.format(accuracy = acc))
    if(acc > 0.98):
        break
```
